```python
### imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
### Algo
def nfl_qb_scrape_tool(URL):

    ### Gather URL
    URL = URL
    response = requests.get(URL)
    html = response.text
    ### Parse HTML Contents
    soup = BeautifulSoup(html, "html.parser")
    ### Set rows into list of lists
    nfl_df_raw = []
    for tr in soup.find_all('tr'):
        cols = []
        for td in tr.find_all(['td', 'th']):
            td_text = td.get_text(strip=True)
            if len(td_text):
                cols.append(td_text)
        nfl_df_raw.append(cols)
    #removes empty lines
    #nfl_df_raw = filter(None, nfl_df_raw)
    ### Delete first + duplicate list
    #del nfl_df_raw[0]
    #del nfl_df_raw[21]
    if len(nfl_df_raw) < 45:
        del nfl_df_raw[0]
        del nfl_df_raw[21]
    elif len(nfl_df_raw) >= 45:
        del nfl_df_raw[0]
        del nfl_df_raw[21]
        del nfl_df_raw[41]
    else: 
        del nfl_df_raw[0]
    # Remove "@"
    # Removing element from list of lists
    N = '@'
    nfl_df_raw = [[ele for ele in sub if ele != N] for sub in nfl_df_raw] 
    ### Unzip list
    nfl_2019_qb_rank, nfl_2019_qb_player, nfl_2019_qb_pos, nfl_2019_qb_age, nfl_2019_qb_date, nfl_2019_qb_lgthrw, nfl_2019_qb_tm, nfl_2019_qb_opp, nfl_2019_qb_result, nfl_2019_qb_game, nfl_2019_qb_week, nfl_2019_qb_day, nfl_2019_qb_cmp, nfl_2019_qb_att, nfl_2019_qb_cmpper, nfl_2019_qb_yds, nfl_2019_qb_td, nfl_2019_qb_int, nfl_2019_qb_rate, nfl_2019_qb_sk, nfl_2019_qb_ydslost, nfl_2019_qb_ya, nfl_2019_qb_aya  = map(list, zip(*nfl_df_raw))
    ### Merge into df
    nfl_2019_qb_df = pd.DataFrame(
        {'Rank': nfl_2019_qb_rank,
         'Player': nfl_2019_qb_player,
         'Position': nfl_2019_qb_pos, 
         'Age': nfl_2019_qb_age, 
         'Date': nfl_2019_qb_date, 
         'League': nfl_2019_qb_lgthrw, 
         'Team': nfl_2019_qb_tm, 
         'Opp': nfl_2019_qb_opp, 
         'Result': nfl_2019_qb_result, 
         'Game': nfl_2019_qb_game, 
         'Week': nfl_2019_qb_week, 
         'Day': nfl_2019_qb_day, 
         'Cmp': nfl_2019_qb_cmp, 
         'Pass_Att': nfl_2019_qb_att, 
         'Cmp_Per': nfl_2019_qb_cmpper, 
         'Pass_Yards': nfl_2019_qb_yds, 
         'Pass_TD': nfl_2019_qb_td, 
         'Pass_Int': nfl_2019_qb_int, 
         'QB_Rate': nfl_2019_qb_rate, 
         'Sacked': nfl_2019_qb_sk, 
         'Yds_Lost': nfl_2019_qb_ydslost, 
         'Yards_Att': nfl_2019_qb_ya, 
         'Adj_Yards_Att': nfl_2019_qb_aya
        })
    nfl_2019_qb_df.columns = nfl_2019_qb_df.iloc[0]
    nfl_2019_qb_df = nfl_2019_qb_df.reindex(nfl_2019_qb_df.index.drop(0)).reset_index(drop=True)
    nfl_2019_qb_df.columns = ['Rank', 'Player', 'Position', 'Age', 'Date', 'League', 'Team', 'Opp', 'Result', 'Game', 'Week', 'Day', 'Cmp', 'Pass_Att', 'Cmp_Per', 'Pass_Yards', 'Pass_TD', 'Pass_Int', 'QB_Rate', 'Sacked', 'Yds_Lost', 'Yards_Att', 'Adj_Yards_Att']
    nfl_2019_qb_df.Rank = nfl_2019_qb_df.Rank.astype('category')
    nfl_2019_qb_df.Player = nfl_2019_qb_df.Player.astype('str')
    nfl_2019_qb_df['Position'] = nfl_2019_qb_df['Position'].astype('category')
    nfl_2019_qb_df['Age'] = nfl_2019_qb_df['Age'].astype('float')
    nfl_2019_qb_df['Date'] =  pd.to_datetime(nfl_2019_qb_df['Date'])
    nfl_2019_qb_df['League'] = nfl_2019_qb_df['League'].astype('category')
    nfl_2019_qb_df['Team'] = nfl_2019_qb_df['Team'].astype('category')
    nfl_2019_qb_df['Opp'] = nfl_2019_qb_df['Opp'].astype('category')
    nfl_2019_qb_df['Game'] = nfl_2019_qb_df['Game'].astype('category')
    nfl_2019_qb_df['Week'] = nfl_2019_qb_df['Week'].astype('category')
    nfl_2019_qb_df['Day'] = nfl_2019_qb_df['Day'].astype('category')
    nfl_2019_qb_df['Cmp'] = nfl_2019_qb_df['Cmp'].astype('int')
    nfl_2019_qb_df['Pass_Att'] = nfl_2019_qb_df['Pass_Att'].astype('int')
    nfl_2019_qb_df['Cmp_Per'] = nfl_2019_qb_df['Cmp_Per'].astype('float')
    nfl_2019_qb_df['Pass_Yards'] = nfl_2019_qb_df['Pass_Yards'].astype('int')
    nfl_2019_qb_df['Pass_TD'] = nfl_2019_qb_df['Pass_TD'].astype('int')
    nfl_2019_qb_df['Pass_Int'] = nfl_2019_qb_df['Pass_Int'].astype('int')
    nfl_2019_qb_df['QB_Rate'] = nfl_2019_qb_df['QB_Rate'].astype('float')
    nfl_2019_qb_df['Sacked'] = nfl_2019_qb_df['Sacked'].astype('int')
    nfl_2019_qb_df['Yds_Lost'] = nfl_2019_qb_df['Yds_Lost'].astype('int')
    nfl_2019_qb_df['Yards_Att'] = nfl_2019_qb_df['Yards_Att'].astype('float')
    nfl_2019_qb_df['Adj_Yards_Att'] = nfl_2019_qb_df['Yards_Att'].astype('float')
    return nfl_2019_qb_df
```
