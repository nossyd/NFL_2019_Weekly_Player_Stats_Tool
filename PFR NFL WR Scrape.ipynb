```python
### imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
### Algo
def nfl_wr_scrape_tool(URL):

    ### Gather URL
    URL = URL
    response = requests.get(URL)
    html = response.text
    ### Parse HTML Contents
    soup = BeautifulSoup(html, "html.parser")
    ### Set rows into list of lists
    nfl_df_raw = []
    for tr in soup.find_all('tr'):
        cols = []
        for td in tr.find_all(['td', 'th']):
            td_text = td.get_text(strip=True)
            if len(td_text):
                cols.append(td_text)
        nfl_df_raw.append(cols)
    ### Delete first + duplicate list
    del nfl_df_raw[0]
    del nfl_df_raw[21]
    del nfl_df_raw[41]
    del nfl_df_raw[61]
    if len(nfl_df_raw) > 80:
        del nfl_df_raw[81]
    # Remove "@"
    # Removing element from list of lists
    N = '@'
    nfl_df_raw = [[ele for ele in sub if ele != N] for sub in nfl_df_raw] 
    ### Unzip list
    nfl_2019_wr_rank, nfl_2019_wr_player, nfl_2019_wr_pos, nfl_2019_wr_age, nfl_2019_wr_date, nfl_2019_wr_lg, nfl_2019_wr_tm, nfl_2019_wr_opp, nfl_2019_wr_result, nfl_2019_wr_game, nfl_2019_wr_week, nfl_2019_wr_day, nfl_2019_wr_trgt, nfl_2019_wr_rec, nfl_2019_wr_yds, nfl_2019_wr_ypr, nfl_2019_wr_td, nfl_2019_wr_ctchper, nfl_2019_wr_ydspertrgt = map(list, zip(*nfl_df_raw))
    ### Merge into df
    nfl_2019_wr_df = pd.DataFrame(
        {'Rank': nfl_2019_wr_rank,
         'Player': nfl_2019_wr_player,
         'Position': nfl_2019_wr_pos, 
         'Age': nfl_2019_wr_age, 
         'Date': nfl_2019_wr_date, 
         'League': nfl_2019_wr_lg, 
         'Team': nfl_2019_wr_tm, 
         'Opp': nfl_2019_wr_opp, 
         'Result': nfl_2019_wr_result, 
         'Game': nfl_2019_wr_game, 
         'Week': nfl_2019_wr_week, 
         'Day': nfl_2019_wr_day,    
         'Targets': nfl_2019_wr_trgt, 
         'Recs': nfl_2019_wr_rec, 
         'Rec_Yards': nfl_2019_wr_yds, 
         'Yards_per_Rec': nfl_2019_wr_ypr, 
         'Rec_TD': nfl_2019_wr_td, 
         'Catch_Percentage': nfl_2019_wr_ctchper, 
         'Yard_per_Target': nfl_2019_wr_ydspertrgt, 
        })
    nfl_2019_wr_df.columns = nfl_2019_wr_df.iloc[0]
    nfl_2019_wr_df = nfl_2019_wr_df.reindex(nfl_2019_wr_df.index.drop(0)).reset_index(drop=True)
    nfl_2019_wr_df.columns = ['Rank', 'Player', 'Position', 'Age', 'Date', 'League', 'Team', 'Opp', 'Result', 'Game', 'Week', 'Day', 'Targets', 'Recs', 'Rec_Yards', 'Yards_per_Rec', 'Rec_TD', 'Catch_Percentage', 'Yard_per_Target']
    nfl_2019_wr_df.Rank = nfl_2019_wr_df.Rank.astype('category')
    nfl_2019_wr_df.Player = nfl_2019_wr_df.Player.astype('str')
    nfl_2019_wr_df['Position'] = nfl_2019_wr_df['Position'].astype('category')
    nfl_2019_wr_df['Age'] = nfl_2019_wr_df['Age'].astype('float')
    nfl_2019_wr_df['Date'] =  pd.to_datetime(nfl_2019_wr_df['Date'])
    nfl_2019_wr_df['League'] = nfl_2019_wr_df['League'].astype('category')
    nfl_2019_wr_df['Team'] = nfl_2019_wr_df['Team'].astype('category')
    nfl_2019_wr_df['Opp'] = nfl_2019_wr_df['Opp'].astype('category')
    nfl_2019_wr_df['Game'] = nfl_2019_wr_df['Game'].astype('category')
    nfl_2019_wr_df['Week'] = nfl_2019_wr_df['Week'].astype('category')
    nfl_2019_wr_df['Day'] = nfl_2019_wr_df['Day'].astype('category')
    nfl_2019_wr_df['Targets'] = nfl_2019_wr_df['Targets'].astype('int')
    nfl_2019_wr_df['Recs'] = nfl_2019_wr_df['Recs'].astype('int')
    nfl_2019_wr_df['Rec_Yards'] = nfl_2019_wr_df['Rec_Yards'].astype('int')
    nfl_2019_wr_df['Yards_per_Rec'] = nfl_2019_wr_df['Yards_per_Rec'].astype('float')
    nfl_2019_wr_df['Rec_TD'] = nfl_2019_wr_df['Rec_TD'].astype('int')
    # Remove "%"
    nfl_2019_wr_df['Catch_Percentage'] = [sub.replace('%', '') for sub in nfl_2019_wr_df['Catch_Percentage']] 
    nfl_2019_wr_df['Catch_Percentage'] = nfl_2019_wr_df['Catch_Percentage'].astype('float')
    nfl_2019_wr_df['Yard_per_Target'] = nfl_2019_wr_df['Yard_per_Target'].astype('float')
    return nfl_2019_wr_df
```
