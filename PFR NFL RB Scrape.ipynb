```python
### imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
### Algo
def nfl_rb_scrape_tool(URL):

    ### Gather URL
    URL = URL
    response = requests.get(URL)
    html = response.text
    ### Parse HTML Contents
    soup = BeautifulSoup(html, "html.parser")
    ### Set rows into list of lists
    nfl_df_raw = []
    for tr in soup.find_all('tr'):
        cols = []
        for td in tr.find_all(['td', 'th']):
            td_text = td.get_text(strip=True)
            if len(td_text):
                cols.append(td_text)
        nfl_df_raw.append(cols)
    ### Delete first + duplicate list
    del nfl_df_raw[0]
    del nfl_df_raw[21]
    del nfl_df_raw[41]
    del nfl_df_raw[61]
    if len(nfl_df_raw) > 80:
        del nfl_df_raw[81]
    # Remove "@"
    # Removing element from list of lists
    N = '@'
    nfl_df_raw = [[ele for ele in sub if ele != N] for sub in nfl_df_raw] 
    ### Unzip list
    nfl_2019_rb_rank, nfl_2019_rb_player, nfl_2019_rb_pos, nfl_2019_rb_age, nfl_2019_rb_date, nfl_2019_rb_lg, nfl_2019_rb_tm, nfl_2019_rb_opp, nfl_2019_rb_result, nfl_2019_rb_game, nfl_2019_rb_week, nfl_2019_rb_day, nfl_2019_rb_att, nfl_2019_rb_yds, nfl_2019_rb_yds_att, nfl_2019_rb_td = map(list, zip(*nfl_df_raw))
    ### Merge into df
    nfl_2019_rb_df = pd.DataFrame(
        {'Rank': nfl_2019_rb_rank,
         'Player': nfl_2019_rb_player,
         'Position': nfl_2019_rb_pos, 
         'Age': nfl_2019_rb_age, 
         'Date': nfl_2019_rb_date, 
         'League': nfl_2019_rb_lg, 
         'Team': nfl_2019_rb_tm, 
         'Opp': nfl_2019_rb_opp, 
         'Result': nfl_2019_rb_result, 
         'Game': nfl_2019_rb_game, 
         'Week': nfl_2019_rb_week, 
         'Day': nfl_2019_rb_day,    
         'Attempts': nfl_2019_rb_att, 
         'Rush_Yards': nfl_2019_rb_yds, 
         'Yards_per_Attempt': nfl_2019_rb_yds_att, 
         'Rush_TD': nfl_2019_rb_td 
        })
    nfl_2019_rb_df.columns = nfl_2019_rb_df.iloc[0]
    nfl_2019_rb_df = nfl_2019_rb_df.reindex(nfl_2019_rb_df.index.drop(0)).reset_index(drop=True)
    nfl_2019_rb_df.columns = ['Rank', 'Player', 'Position', 'Age', 'Date', 'League', 'Team', 'Opp', 'Result', 'Game', 'Week', 'Day', 'Attempts', 'Rush_Yards', 'Yards_per_Attempt', 'Rush_TD']
    nfl_2019_rb_df.Rank = nfl_2019_rb_df.Rank.astype('category')
    nfl_2019_rb_df.Player = nfl_2019_rb_df.Player.astype('str')
    nfl_2019_rb_df['Position'] = nfl_2019_rb_df['Position'].astype('category')
    nfl_2019_rb_df['Age'] = nfl_2019_rb_df['Age'].astype('float')
    nfl_2019_rb_df['Date'] =  pd.to_datetime(nfl_2019_rb_df['Date'])
    nfl_2019_rb_df['League'] = nfl_2019_rb_df['League'].astype('category')
    nfl_2019_rb_df['Team'] = nfl_2019_rb_df['Team'].astype('category')
    nfl_2019_rb_df['Opp'] = nfl_2019_rb_df['Opp'].astype('category')
    nfl_2019_rb_df['Game'] = nfl_2019_rb_df['Game'].astype('category')
    nfl_2019_rb_df['Week'] = nfl_2019_rb_df['Week'].astype('category')
    nfl_2019_rb_df['Day'] = nfl_2019_rb_df['Day'].astype('category')
    nfl_2019_rb_df['Attempts'] = nfl_2019_rb_df['Attempts'].astype('int')
    nfl_2019_rb_df['Rush_Yards'] = nfl_2019_rb_df['Rush_Yards'].astype('int')
    nfl_2019_rb_df['Yards_per_Attempt'] = nfl_2019_rb_df['Yards_per_Attempt'].astype('float')
    nfl_2019_rb_df['Rush_TD'] = nfl_2019_rb_df['Rush_TD'].astype('int')
    return nfl_2019_rb_df
```
